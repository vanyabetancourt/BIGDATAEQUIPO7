```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

from sklearn.decomposition import TruncatedSVD

from google.colab import drive
drive.mount('/content/drive')

DIR = "/content/drive/MyDrive/Colab Notebooks/MNA/Big Data/Modulo 2 Sistema de Recomendacion/"
os.chdir(DIR)
```

```python
# Lee los dos archivos que contienen el listado de bebidas y los ratings dado por los usuarios
data1 = pd.read_csv("starbucks_ratings.csv", header='infer', sep=",")
data2 = pd.read_csv("starbucks_beverages.csv", header='infer')

print(data1.shape, data2.shape)
```

```python
#Del primer archivo obtenemos los datos con las evaluaciones de las bebidas dadas por los usuarios
#Las evaluaciones pueden ser 0, 1 o 2, siendo 0 la menor calificación y 2 la mayor calificación

data1.head()
```

```python
#Del segundo archivo obtenemos el catalogo con el nombre de las bebidas
data2.head()
```

```python
# Hacemon un dataframe conmbianado para tener en un mismo dataset los ratings y nombres de las bebidas

df_combinado = pd.merge(data1,data2, on='Beverage')

df_combinado.head()
```

```python
#    Define la matriz de utilidad cuyos renglones sean los nombres de las
#    bebidas, las columnas los IDs de los usuarios y los ratings de la
#    evaluación de las bebidas (rating). La llamaremos "beverage_ratings".

beverage_ratings = df_combinado.pivot_table(values='rating',index='Beverage name',columns='userID',fill_value=0)

print('Dimensión de la matriz de Utilidad sobre la evaluación de las bebidas:')
print('(bebidas, usuarios) =', (beverage_ratings.shape))

beverage_ratings.head()
```

```python
# Apliquemos la transformación SVD para obtener la matriz de las variables
# latentes de las bebidas en relación a la evalución de los usuarios con
# respecto a su sabor. Deberás obtener la factorización SVD con el máximo de
# componentes y por lo tanto el máximo de valores singulares posible.

nc_bev = 130

SVD_bev = TruncatedSVD( n_components = nc_bev )
SVD_bev.fit(beverage_ratings)

# Determinemos la cantidad de componentes que explican un 90%
# de la variabilidad acumulada de cada componente:

for j in range(nc_bev):
  if SVD_bev.explained_variance_ratio_[0:j].sum() > 0.90:
    break

# Usaremos esta cantidad de componentes para las recomendaciones basadas
# en la calificación de las bebidas:
N_bev = j-1

print('Total de valores singulares basados en evaluación de la :', nc_bev)
print('Valor de truncamiento al 90% de dicha variabilidad:',N_bev)
```

```python
# Usando la métrica de correlación de Pearson, obtendremos las 10 mejores
# recomendaciones de este modelo no supervisado con base a la información de
# alguien que desea obtener las similidades con la bebida "Classic Espresso Drinks-Coffee Americano-Tall"
# Para ello obtendremos las correlaciones positivas y de ahí deberás seleccionar
# las diez mejores.

# Factorización SVD:
SVD_bev = TruncatedSVD(n_components = N_bev)
resultant_matrix_bev = SVD_bev.fit_transform(beverage_ratings)

# Matriz de correlación de Pearson:
corr_mat_bev = np.corrcoef(resultant_matrix_bev)

# Restaurante de referencia:
bebida_de_referencia = "Classic Espresso Drinks-Coffee Americano-Tall"
nombres_rest = beverage_ratings.T.columns  # nombres de restaurantes
idx_rest = list(nombres_rest).index(bebida_de_referencia)
corr_rest = corr_mat_bev[idx_rest] # Vector de Correlación del RinconHuasteco

# Buscando las correlaciones positivas:
idx = (corr_rest>0)
mejores_sim_bev = list()
for i in range(len(nombres_rest[idx])):
  mejores_sim_bev.append((corr_rest[idx][i], nombres_rest[idx][i]))

print('Total de similaridades positivas encontradas:', len(mejores_sim_bev))

print('Algunos de los resultados encontrados:')
mejores_sim_bev[0:10]
```
